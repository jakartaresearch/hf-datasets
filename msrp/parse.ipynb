```python
from tqdm import tqdm
import pandas as pd
```

```python
with open('msr_paraphrase_train.txt', 'r', encoding='utf8') as f:
    train = f.read()
```

```python
with open('msr_paraphrase_test.txt', 'r', encoding='utf8') as f:
    test = f.read()
```

### Train

```python
train_list = train.split('\n')
```

```python
col_train = train_list.pop(0)
```

```python
col = ['Quality', '#1 ID', '#2 ID', '#1 String', '#2 String']
```

```python
parsed_train_list = []
for row in tqdm(train_list):
    parsed_train_list.append(row.split('\t'))
```

```python
parsed_train_list
```

### Test

```python
test_list = test.split('\n')
```

```python
_ = test_list.pop(0)
```

```python
test_list[0]
```

```python
parsed_test_list = []
for row in test_list:
    parsed_test_list.append(row.split('\t'))
```

```python
parsed_test_list
```

### Make dataset

```python
d_train = pd.DataFrame(parsed_train_list, columns=col)
d_test = pd.DataFrame(parsed_test_list, columns=col)
```

```python
d_train.to_csv('train.csv', index=False)
```
