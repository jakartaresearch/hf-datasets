```python
import urllib3
from bs4 import BeautifulSoup
```

```python
http = urllib3.PoolManager()
```

```python
res = http.request("GET", "https://reviews.femaledaily.com/products/treatment/serum-essence-23/the-ordinary/the-ordinary-niacinamide-10-zinc-1?cat=&cat_id=0&age_range=&skin_type=&skin_tone=&skin_undertone=&hair_texture=&hair_type=&order=newest&page=1")
```

```python
res.status
```

```python
soup = BeautifulSoup(res.data, "lxml")
```

```python
soup
```

```python
review_box = soup.find('div', attrs={'class': 'jsx-584641148 jsx-134561806 list-reviews'})
```

```python
reviews = review_box.find_all('div', attrs={'class': 'jsx-584641148 jsx-134561806 item'})
```

```python
for review in reviews:
    print(review.find('p', attrs={'class': 'text-content'}))
    print('\n')
```
